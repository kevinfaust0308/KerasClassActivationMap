# Other possible things: 

## Writing predictions on overlay 

```python
# example with multi stacked cam

i = 'data/train/256x256_size/cat/object-102.jpg'
classes = ['cat', 'dog', 'plant']
threshold = 0.4 # only show CAM if a classes' probability is greater than 40%
pretty_top_predictions = True # default. labels predictions
cam, pred = get_multi_stacked_cam(model, i, classes, -4, threshold=threshold, pretty_top_predictions=pretty_top_predictions)
```

```python
# change pred to list
my_list = []
for obj in pred:
    # make pred accuracy 2 decimals
    p = '{:.2f}'.format(pred[obj][0])
    # our new list will have object name, acc, and color label
    my_list.append([obj] + [p] + [pred[obj][1]])
my_list
```

```python
text_color = (0,255,255)

# starting top offset
top_offset = 30

# loop through all those pred labels and add to image
for label in my_list:

    cv2.putText(
        cam, # put text on this image
        "{}".format(label), # text
        (10,top_offset),# left offset, top offset
        cv2.FONT_HERSHEY_SIMPLEX, # font family
        0.7, # scale of text size
        text_color, # color
        1, # line width
        cv2.LINE_AA # line type
    )
    
    # move next text down for spacing
    top_offset += 30

# cv2.imshow('CAM', cam)
# cv2.waitKey(0)
```

```python
plt.imshow(cam)
```

## Writing max prediction on overlay 

```python
i = 'data/train/256x256_size/dog/object-154.jpg'
classes = ['cat', 'dog', 'plant']
show_top_x_classes = 1 # only show CAM if a classes' probability is greater than 40%
pretty_top_predictions = True # default. labels predictions
cam, pred = get_multi_stacked_cam(model, i, classes, -4, show_top_x_classes=show_top_x_classes, pretty_top_predictions=pretty_top_predictions)
```

```python
max_pred = "Object", 0, 'Color'
# loop through predictions
for k in pred:
    # update max pred if current pred is bigger than max
    if pred[k][0] > max_pred[1]:
        max_pred = k, pred[k][0], pred[k][1]
max_pred
```

```python
text_color = (0,0,0)

name, acc, color  = max_pred[0], max_pred[1], max_pred[2]

cv2.putText(
    cam, # put text on this image
    "Label: {}".format(name), # text
    (10,30),# left offset, top offset
    cv2.FONT_HERSHEY_SIMPLEX, # font family
    0.7, # scale of text size
    text_color, # color
    1, # line width
    cv2.LINE_AA # line type
)

cv2.putText(
    cam, # put text on this image
    "Confidence: {:.2f}".format(acc*100), # text two decimal places (https://mkaz.tech/code/python-string-format-cookbook/)
    (10,55),# left offset, top offset
    cv2.FONT_HERSHEY_SIMPLEX, # font family
    0.7, # scale of text size
    text_color, # color
    1, # line width
    cv2.LINE_AA # line type
)

cv2.putText(
    cam, # put text on this image
    "Color: {}".format(color),
    (10,80),# left offset, top offset
    cv2.FONT_HERSHEY_SIMPLEX, # font family
    0.7, # scale of text size
    text_color, # color
    1, # line width
    cv2.LINE_AA # line type
)

# cv2.imshow('CAM', cam)
# cv2.waitKey(0)
```
