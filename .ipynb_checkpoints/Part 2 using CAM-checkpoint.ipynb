# generating CAM on different images

```python
# load model
model = load_model('VGG19_trained.h5')
```

```python
# run custom function on a cat
cam, pred = predict_label_with_cam(model, 'test_cat_img.jpg', -4, overlay=True, overlay_alpha=0.6)
```

```python
# run custom function on a plant
cam, pred = predict_label_with_cam(model, 'test_plant_img.jpg', -4, overlay=True, overlay_alpha=0.5)
```

```python
plt.imshow(cam)
```

```python
pred
```

# writing prediction on overlay 

```python
# get max prediction
i = np.argmax(pred)
print(i)
pred = pred[0][i]
print(pred)

# index 2 is plant

name = 'Plant'
```

```python
text_color = (255,255,255)

cv2.putText(
    cam, # put text on this image
    "Label: {}".format(name), # text
    (10,30),# left offset, top offset
    cv2.FONT_HERSHEY_SIMPLEX, # font family
    0.8, # scale of text size
    text_color, # color
    2, # line width
    cv2.LINE_AA # line type
)

cv2.putText(
    cam, # put text on this image
    "Confidence: {:.2f}".format(pred*100), # text two decimal places (https://mkaz.tech/code/python-string-format-cookbook/)
    (10,55),# left offset, top offset
    cv2.FONT_HERSHEY_SIMPLEX, # font family
    0.7, # scale of text size
    text_color, # color
    2, # line width
    cv2.LINE_AA # line type
)

cv2.imshow('CAM', cam)
cv2.waitKey(0)
```
