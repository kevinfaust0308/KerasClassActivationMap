# array of all the cat images

```python
cats = []
cat_file_paths = glob.glob('data/train/cats/cat*')
print(len(cat_file_paths))
for fp in cat_file_paths:
    temp = misc.imread(fp)
    cats.append(misc.imread(fp))

cats = np.asarray(cats) 
cats.shape
```

# array of all the dog images

```python
dogs = []
dog_file_paths = glob.glob('data/train/dog*')
print(len(dog_file_paths))
for fp in dog_file_paths:
    dogs.append(misc.imread(fp))

dogs = np.asarray(dogs) 
dogs.shape
```

# Create labels for each category

```python
cats_label = np.zeros(cats.shape)
dogs_label = np.ones(dogs.shape)
```

# create X and y matrix

```python
X = np.concatenate((cats, dogs))
y = np.concatenate((cats_label, dogs_label))
```

# max width/height of our images

```python
# for 3d, (width, height, channel)
dict_w = {}
dict_h = {}

for img in X:
    w, h ,c = img.shape
    
    if w in dict_w:
        dict_w[w] += 1
    else:
        dict_w[w]=1
    if h in dict_h:
        dict_h[h]+=1
    else:
        dict_h[h]=1
        
    
print('most common height', max(dict_h, key=dict_h.get))
print('biggest height', max(dict_h.keys()), '\n')
# dict_h
print('most common width', max(dict_w, key=dict_w.get))
print('biggest width', max(dict_w.keys()))
# dict_w
```

# image stuff

# doing flow_from_directory. otherwise too slow

```python
batch_size = 16

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
        rescale=1./255,
#         rotation_range=40,
#         width_shift_range=0.2,
#         height_shift_range=0.2,
#         shear_range=0.2,
#         zoom_range=0.2,
#         horizontal_flip=True,
#         vertical_flip=True,
#         fill_mode='nearest'
)


# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(rescale=1./255)

# this is a generator that will read pictures found in
# subfolers of 'data/train', and indefinitely generate
# batches of augmented image data
train_generator = train_datagen.flow_from_directory(
        'data/train',  # this is the target directory
        target_size=(500, 435),  # all images will be resized
        batch_size=batch_size,
        class_mode='binary')  # since we use binary_crossentropy loss, we need binary labels

# this is a similar generator, for validation data
validation_generator = test_datagen.flow_from_directory(
        'data/validation',
        target_size=(500, 435),
        batch_size=batch_size,
        class_mode='binary')


# # generate 20 augmented images of a random pic
# i=0
# # resize picture to be 4d. we want (samples, rows, cols, channels)
# temp = X[13].reshape((1,) + X[13].shape)
# for batch in train_datagen.flow(x=temp, batch_size=1, save_to_dir='data/augmented_img_preview', save_prefix='cat'):
#     i+=1
#     if i > 20:
#         break
```

```python
# create the model
augmented_model = Sequential()
augmented_model.add(Conv2D(32, (3,3), padding='same', input_shape=(500,435,3), activation='relu', kernel_constraint=maxnorm(3)))
augmented_model.add(Dropout(0.2))
augmented_model.add(Conv2D(32, (3,3), padding='same', activation='relu', kernel_constraint=maxnorm(3)))
augmented_model.add(MaxPooling2D())
augmented_model.add(Flatten())
augmented_model.add(Dense(512, activation='relu', kernel_constraint=maxnorm(3)))
augmented_model.add(Dropout(0.5))
augmented_model.add(Dense(1, activation='sigmoid'))

# compile the model
augmented_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = augmented_model.fit_generator(train_generator, 
                    steps_per_epoch= 20000 // batch_size,
                    epochs=1,
                    validation_data=validation_generator,
                    validation_steps=5000 // batch_size)
```
